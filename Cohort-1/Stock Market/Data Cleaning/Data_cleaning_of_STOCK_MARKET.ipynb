```python
import yfinance as yf
import pandas as pd
tickers=['AAPL','MSFT','SPY']
data = yf.download(tickers, period="5y", interval="1d", auto_adjust=True, progress=False)
data.columns = [f"{ticker}_{metric}" for metric, ticker in data.columns]
data = data.reset_index()
data
```

```python
import numpy as np
price_cols = [col for col in data.columns if 'Close' in col or 'Open' in col or 'High' in col or 'Low' in col]
for col in price_cols:
    data[col] = data[col].apply(lambda x: np.nan if x <= 0 else x)
data[price_cols] = data[price_cols].ffill()
```

```python
import pandas as pd
data = data.reset_index()
data['Date'] = pd.to_datetime(data['Date'])
```
